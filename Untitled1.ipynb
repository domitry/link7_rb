```ruby
require 'ode'
```

```ruby
f = Proc.new{|t, y, fargs|
  p, r, b = fargs[:p], fargs[:r], fargs[:b]
    [ -p*y[0] + p*y[1],
      -y[0]*y[2] + r*y[0] - y[1],
      y[0]*y[1] - b*y[2]
    ]
}

t=0.0; result = []
solver = Ode::Solver.new(f).init(0, [0.1, 0.1, 0.1]).f_args({p: 10, r: 28, b: 8/3})
(30000).times do
  t += 0.001
  solver.integrate(t)
result.push(solver.y)
end
```
